```python
q=[3, 4, 7, 8]
d_ans = {}
d_tmp = {}
for num in q:
    print('======================')
    print(f'num={num}')
    print(f'd_ans={d_ans}')
    if len(d_ans)==0: ### 要素がない場合は、登録するだけ
        #print('hoge')
        d_ans[f'{num}']=num
        continue
    #print('---key---')
    #print(next(iter(d_ans))) ### 辞書から要素を一つ取り出す
    #print('---key---')
    ### 4つの演算を行い、それぞれを辞書に登録
    #for key in next(iter(d_ans)):
    for key in list(d_ans.keys()):
        print(f'key={key}')
        value = d_ans.pop(key) ### 値を取得して、レコードを削除
        #print(f'value={value}')
        d_tmp[f'{key}+{num}'] = value + num
        d_tmp[f'{key}-{num}'] = value - num
        d_tmp[f'{key}*{num}'] = value * num
        if num!=0: d_tmp[f'{key}/{num}'] = value / num ### 0では割らない
        #print(f'd_tmp={d_tmp}')
    ### 辞書の中身がなくなったら、結果をコピーして、次の数へ進む
    d_ans = d_tmp.copy()
    d_tmp = {}
    print(f'd_ans_final={d_ans}')
```

```python
###################################################
### 関数：数値リストを受け取り、四則演算を総当たりする。
### 入力：数値リスト
### 出力：辞書（キー：演算式、バリュー：演算結果）
###################################################
def calc_all_pattern(q):
    d_ans = {}
    d_tmp = {}
    for num in q:
        #print('======================')
        #print(f'num={num}')
        #print(f'd_ans={d_ans}')
        if len(d_ans)==0: ### 要素がない場合は、登録するだけ
            #print('hoge')
            d_ans[f'{num}']=num
            continue
        #print('---key---')
        #print(next(iter(d_ans))) ### 辞書から要素を一つ取り出す
        #print('---key---')
        ### 4つの演算を行い、それぞれを辞書に登録
        #for key in next(iter(d_ans)):
        for key in list(d_ans.keys()):
            #print(f'key={key}')
            value = d_ans.pop(key) ### 値を取得して、レコードを削除
            #print(f'value={value}')
            d_tmp[f'{key}+{num}'] = value + num
            d_tmp[f'{key}-{num}'] = value - num
            d_tmp[f'{key}--{num}'] = num - value   ### 掛け算や割り算のあとに、どちらから引くかで答えが変わる
            d_tmp[f'{key}*{num}'] = value * num
            if num!=0: d_tmp[f'{key}/{num}'] = value / num ### 0では割らない
            #print(f'd_tmp={d_tmp}')
        ### 辞書の中身がなくなったら、結果をコピーして、次の数へ進む
        d_ans = d_tmp.copy()
        d_tmp = {}
        #print(f'd_ans_final={d_ans}')
    return d_ans
```

```python
q=[3, 4, 7, 8]
calc_all_pattern(q)
```

```python
d_ans_all={}
for q in q_perm:
    d_ans_all.update(calc_all_pattern(q))
print(len(d_ans_all))
```

```python
### バリューの値から、キーを逆引きする
keys = [k for k, v in d_ans_all.items() if v == 10]
print(keys)
```

# 完成：全体の流れ実行

```python
###### 設定 ###############################
q = [3,4,7,8] ### 問題
target = 10   ### 探したい答え
##########################################
print('question:' + str(q))
print('target:' + str(target))
d_ans_all={}

### 総当たりで演算を行う
for q in q_perm:
    d_ans_all.update(calc_all_pattern(q))
print('all pattern : ' + str(len(d_ans_all)))

### バリューの値から、キーを逆引きする
keys = [k for k, v in d_ans_all.items() if v == target]
print('correct pattern : ' + str(len(keys)))
print(keys)
```
